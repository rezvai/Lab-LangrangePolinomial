### Import libraries

```python
from sympy import symbols, expand
import sympy
import matplotlib.pyplot as plt
```

________________________________________________________________________________

## 1. First excercise 

* 1.1 Getting the Polynomial function

```python
x = symbols('x')

nums = [2, 3, 4]

a = 2.5

f = lambda x: sympy.ln(x) ** (13/4)

urav_main = 0
for i in range(len(nums)):
    urav_high = 1
    urav_low = 1
    for j in range(len(nums)):
        if j != i:
            urav_high *= (x-nums[j])
            urav_low *= (nums[i]-nums[j])
    urav_main += f(nums[i]) * (urav_high / urav_low)
        
simplified = expand(urav_main)

urav_main
```

* 1.2 Printing the Answer of Approximated Function

```python
print('Answer of the approximated function:', f(a))
```

* 1.3 Printing the Simplified Expression

```python
print('Simplified expression of Langrange polynomial:')
simplified
```

* 1.4 Printing the Answer of Approximated Function (Polynomial)

```python
polynomial_ans = simplified.subs(x, a).evalf()
print('Approximated answer at a:', polynomial_ans)
```

* 1.5 Printing the Loss Between Approximated Function and Polynomial Function

```python
R = abs(f(a) - polynomial_ans)
print('Loss:', R)
```

* 1.6 Printing Graphs for Nodes

* * 1.6.1 Plot of the Approximated Function

```python
plt.figure(figsize=(10, 6))
plt.plot(nums, [f(i) for i in nums], label='Approximated Function', color='blue')
plt.plot(nums, [simplified.subs(x, i) for i in nums], label='Langrage Polynomial', color='red', linestyle='--')
plt.scatter(nums, [f(i) for i in nums], color='green', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

* * 1.6.2 Plot of the Langrange Polynomial

```python
plt.figure(figsize=(10, 6))
plt.plot(nums, [simplified.subs(x, i) for i in nums], label='Langrage Polynomial', color='blue')
plt.plot(nums, [f(i) for i in nums], label='Approximated Function', color='red', linestyle='--')
plt.scatter(nums, [simplified.subs(x, i) for i in nums], color='green', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
